# JOINS IN PANDAS

```python
import pandas as pd
```

```python
raw_data={
    'subject_id':[1,2,3,4,5],
    'first_name':["amit",'sumit','mohan','ravi','kapil'],
    'last_name':['Puir','Sharma','Pathak','Dubey','Nayak']
}
```

```python
df_a=pd.DataFrame(raw_data,columns=["subject_id","first_name",'last_name'])
```

```python
print(df_a)
```

```python
raw_data={
    'subject_id':[4,5,6,7,8],
    'first_name':["Ram",'Shyam','Nisith','Kajal','Harshita'],
    'last_name':['Puir','Sharma','Pathak','Dubey','Nayak']
}

df_b=pd.DataFrame(raw_data,columns=["subject_id","first_name",'last_name'])
print(df_b)
```

```python
df_new=pd.concat([df_a,df_b],ignore_index=True) #ignore_index=True will nake index increase from first data frame
df_new
```

```python
df_new1=pd.concat([df_a,df_b])
df_new1
```

```python
df_new1=pd.concat([df_a,df_b],axis=1)
df_new1
```

```python
df_new1=pd.concat([df_a,df_b],ignore_index=True,axis=1)
df_new1
```
