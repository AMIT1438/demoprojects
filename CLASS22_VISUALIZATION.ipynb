```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6,7])
b=np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()
```

```python
a=np.array([14,26,38,94,95,36,67])
b=np.array([1,2,3,4,5,6,7])
plt.title("Bar Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.bar(a,b)
plt.show()
```

```python
a=np.array([14,26,38,94,95,36,67])
b=np.array([1,2,3,4,5,6,7])
plt.title("Bar Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.bar(a,b, color='g', align='center')
plt.show()
```
