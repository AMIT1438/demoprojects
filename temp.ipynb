```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dsiris=pd.read_csv('iris.csv')
```

```python
print(dsiris)
```

```python
print(dsiris.head())
```

```python
sns.set(style='whitegrid')
```

```python
ax=sns.stripplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
sns.set(style='whitegrid')
ax=sns.swarmplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='class',y='sepal width',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='sepal length',y='class',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.countplot(x='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.catplot(x='class',y='sepal width',data=dsiris)
plt.title('Graph')
plt.show()
```
